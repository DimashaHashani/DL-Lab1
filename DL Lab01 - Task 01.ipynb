```python
import matplotlib.pyplot as plt
```

```python
exp_data = np.random.exponential(scale=1.0, size=(100000,))
uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000)
```

```python
plt.hist(exp_data, density=True, bins=200, histtype="step", color="green", label="Exponential")
plt.hist(uniform_data, density=True, bins=200, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, density=True, bins=200, histtype="step", color="red", label="Normal")

plt.axis([-2.5, 10, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title('Z = X² + Y²')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
from scipy.stats import spearmanr
```

```python
data = {
    'Name': ['Bulbasaur', 'Ivysaur', 'Venusaur', 'Charmander', 'Charmeleon'],
    'Type 1': ['Grass', 'Grass', 'Grass', 'Fire', 'Fire'],
    'Type 2': ['Poison', 'Poison', 'Poison', None, None],
    'Total': [318, 405, 525, 309, 405],
    'HP': [45, 60, 80, 39, 58],
    'Attack': [49, 62, 82, 52, 64],
    'Defense': [49, 63, 83, 43, 58],
    'Sp. Atk': [65, 80, 100, 60, 80],
    'Sp. Def': [65, 80, 100, 50, 65],
    'Speed': [45, 60, 80, 65, 80],
    'Stage': [1, 2, 3, 1, 2],
    'Legendary': [False, False, False, False, False]
}
```

```python
df = pd.DataFrame(data)
```

```python
stats = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
```

```python
pearson_corr = stats.corr(method='pearson')
```

```python
spearman_corr = stats.corr(method='spearman')
```

```python
plt.figure(figsize=(8, 5))
sns.heatmap(pearson_corr, annot=True, cmap='Blues')
plt.title("Pearson Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.heatmap(spearman_corr, annot=True, cmap='viridis', vmin=-1, vmax=1, fmt=".2f")
plt.title("Spearman Rank Correlation Heatmap")
plt.show()
```
