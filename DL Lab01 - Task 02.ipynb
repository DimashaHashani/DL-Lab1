```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Au_nanoparticle_dataset (1).csv')
```

```python
data.head()
```

```python
new_df = data[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

```python
new_df.head(20)
```

```python
print("Mean:\n", new_df.mean())
print("\nStandard Deviation:\n", new_df.std())
print("\nQuartiles:\n", new_df.quantile([0.25, 0.5, 0.75]))
```

```python
plt.figure(figsize=(20, 4))

for i, col in enumerate(new_df.columns):
    plt.subplot(1, 4, i + 1)
    plt.hist(new_df[col], bins=30, color='skyblue', edgecolor='black')
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(new_df)
plt.show()
```

```python
g = sns.PairGrid(new_df)

# Upper triangle: bivariate histogram
g.map_upper(sns.histplot)

# Diagonal: histogram with KDE
g.map_diag(sns.histplot, kde=True)

# Lower triangle: bivariate KDE
g.map_lower(sns.kdeplot, fill=True, cmap='Blues')

plt.show()
```
